# Star Wars Survey

### Overview

```python
import pandas as pd
star_wars = pd.read_csv("additional_files/StarWars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```


***
### Cleaning and Mapping Yes/No Columns

```python
map_dict = {"Yes": True, "No": False}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(map_dict)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(map_dict)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna=False)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False)
```


***
### Cleaning and Mapping Chickbox Columns

```python
import numpy as np
value_map = {"Star Wars: Episode I  The Phantom Menace":True,
            "Star Wars: Episode II  Attack of the Clones":True,
            "Star Wars: Episode III  Revenge of the Sith":True,
            "Star Wars: Episode IV  A New Hope":True,
            "Star Wars: Episode V The Empire Strikes Back":True,
            "Star Wars: Episode VI Return of the Jedi":True,
            np.nan:False}

column_map = {"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
             "Unnamed: 4":"seen_2",
             "Unnamed: 5":"seen_3",
             "Unnamed: 6":"seen_4",
             "Unnamed: 7":"seen_5",
             "Unnamed: 8":"seen_6"}

col_names = star_wars.columns[3:9]
for name in col_names:
    star_wars[name] = star_wars[name].map(value_map)

star_wars = star_wars.rename(columns=column_map)
```

```python
star_wars.head(10)
```


***
### Cleaning the Ranking Columns

```python
star_wars = star_wars.iloc[1:,:]
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float,errors="ignore")
```

```python
star_wars[star_wars.columns[12]].value_counts(dropna=False)
```

```python
rename_map = {"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
             "Unnamed: 10":"ranking_2",
             "Unnamed: 11":"ranking_3",
             "Unnamed: 12":"ranking_4",
             "Unnamed: 13":"ranking_5",
             "Unnamed: 14":"ranking_6"}
star_wars = star_wars.rename(columns=rename_map)
```


***
### Finding the Highest-Ranked Movie

```python
rank_means = star_wars.iloc[1:,9:15].mean(axis=0,skipna=True)
rank_means
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

rank_means.plot.bar(width=.8)
plt.xticks(rotation=45)
plt.show()
```

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As I suspected, it looks like the "original" movies are rated much more highly than the newer ones.



***
### Finding the Most Viewed Movie

```python
seen_sum = star_wars.iloc[:,3:9].sum(axis=0)

seen_sum.plot.bar(width=.8)
plt.xticks(rotation=45)
plt.show()
```

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.



***
### Exploring the Data by Binary Segments (Gender)

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
male_means = males.iloc[1:,9:15].mean(axis=0)
male_means.plot.bar(width=.8)
plt.xticks(rotation=45)
plt.title("Males")
plt.show()

female_means = females.iloc[1:,9:15].mean(axis=0)
female_means.plot.bar(width=.8)
plt.xticks(rotation=45)
plt.title("Female")
plt.show()
```

```python
male_sum = males.iloc[:,3:9].sum(axis=0)
male_sum.plot.bar(width=.8)
plt.xticks(rotation=45)
plt.title("Males")
plt.show()

female_sum = females.iloc[:,3:9].sum(axis=0)
female_sum.plot.bar(width=.8)
plt.xticks(rotation=45)
plt.title("Female")
plt.show()
```



### Male/Femal differences in favorite Star Wars movie and most seen movie

Interestingly, more males watches episodes 1-3, but males liked them far less than females did.
