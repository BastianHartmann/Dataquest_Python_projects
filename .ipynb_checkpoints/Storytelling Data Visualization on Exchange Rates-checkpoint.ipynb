# Storytelling Data Visualization on Exchange Rates

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on <a href=https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020>Kaggle</a> — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on June 2021.

***
### Data Cleaning

Our focus of the project will be on the exchange rate between the euro and the American dollar. Therefore, we focus the data cleaning on the corresponding columns needed.

For data cleaning, we do the following steps:
* We rename the `[US dollar ]` and `Period\Unit:` columns to something easier to type - `US_dollar` and `Time`.
* We change the `Time` column to a `datetime` data type.
* We sort the values by `Time` in ascending order.
* We reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

Furthermore, we isolate the `Time` and the `US_dollar` columns to the DataFrame `euro_to_dollar` and clean missing rows of the `US_dollar` column and convert it to `float` data type.

```python
euro_to_dollar = exchange_rates.loc[:,["Time","US_dollar"]]
bad_dollar_rows = euro_to_dollar["US_dollar"] == "-"
euro_to_dollar = euro_to_dollar[~bad_dollar_rows]
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
euro_to_dollar.info()
```

After the removal of `-` entries in the `US_dollar` column and convertion to the `float` data type, we have `5719` rows left.

***
### Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

Here, we may not want to show the daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the **rolling mean** (also known as the moving average).

We can calculate the mean for each day using the `pandas.Series.rolling().mean()`-method — the only exception is the first day, which doesn't have a previous day. For our rolling mean value, we will use a window length of `30 days`. We add the rolling means to a new column named `rolling_mean`.

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
plt.figure(figsize=(10,6))
plt.plot(euro_to_dollar["Time"],euro_to_dollar["rolling_mean"])
plt.title("Rolling mean of 30 days of the € to US dollar exchange rate")
plt.xlabel("Year")
plt.ylabel("Mean exchange rate of 30 days")
plt.show()
```

In detail, we now want to further investigate how the euro-dollar rate has changed during the coronavirus pandemic. We show the 2020-2021 data and the 2016-2019 data as a baseline.

***
### Coding the Graph

In the following section we write the code to generate the graph, that tells our story of *how the euro-dollar rate has changed during the coronavirus pandemic (2020-2021) compared to 2016 to 2019*

```python
euro_to_dollar_corona = euro_to_dollar[euro_to_dollar["Time"].dt.year >= 2020]
euro_to_dollar_pre_corona = euro_to_dollar[(euro_to_dollar["Time"].dt.year >= 2016)&(euro_to_dollar["Time"].dt.year <= 2019)]

mean_corona = euro_to_dollar_corona["rolling_mean"].mean()
mean_pre_corona = euro_to_dollar_pre_corona["rolling_mean"].mean()
first_year = euro_to_dollar_pre_corona["Time"].min()
second_year = euro_to_dollar_pre_corona["Time"][euro_to_dollar_pre_corona["Time"].dt.year == 2017].min()
third_year = euro_to_dollar_pre_corona["Time"][euro_to_dollar_pre_corona["Time"].dt.year == 2018].min()
fourth_year = euro_to_dollar_pre_corona["Time"][euro_to_dollar_pre_corona["Time"].dt.year == 2019].min()
fivth_year = euro_to_dollar_corona["Time"].min()
sixth_year = euro_to_dollar_corona["Time"][euro_to_dollar_corona["Time"].dt.year == 2021].min()

style.use("fivethirtyeight")

fig = plt.figure(figsize=(12,8))
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,(3,4))
plt.subplots_adjust(hspace=.5)


ax1.plot(euro_to_dollar_pre_corona["Time"],euro_to_dollar_pre_corona["rolling_mean"],color="blue",linewidth=2)
ax1.set_ylim(1.05,1.25)
ax1.set_yticks([1.10,1.15,1.20,1.25])
ax1.set_xticks([first_year, second_year, third_year, fourth_year, fivth_year])
ax1.set_xticklabels(["2016","2017","2018","2019","2020"])
for tick in [1.053,1.10,1.15,1.20,1.247]:
    ax1.axhline(y=tick,color="grey",linewidth=0.5)
ax1.text(17100,1.28,"Pre-Corona (2016 - 2019)",color="blue",weight="bold")
ax1.text(17230,1.32,"EURO-USD rate 30 days average for the years 2016 - 2021",color="black",weight="bold",size=18)
    
ax2.plot(euro_to_dollar_corona["Time"],euro_to_dollar_corona["rolling_mean"],color="red",linewidth=2)
ax2.set_ylim(1.05,1.25)
ax2.set_yticks([1.10,1.15,1.20,1.25])
ax2.set_xticks([])
ax2.set_xticks([fivth_year,sixth_year])
ax2.set_xticklabels(["2020","2021"])
for tick in [1.053,1.10,1.15,1.20,1.247]:
    ax2.axhline(y=tick,color="grey",linewidth=0.5)
ax2.text(18360,1.28,"Corona (2020 - 2021)",color="red",weight="bold")

ax3.plot(euro_to_dollar_pre_corona["Time"],euro_to_dollar_pre_corona["rolling_mean"],color="blue",linewidth=2)
ax3.plot(euro_to_dollar_corona["Time"],euro_to_dollar_corona["rolling_mean"],color="red",linewidth=2)
ax3.set_ylim(1.05,1.25)
ax3.set_yticks([1.10,1.15,1.20,1.25])
ax3.set_xticks([])
ax3.set_xticks([first_year, second_year, third_year, fourth_year, fivth_year,sixth_year])
ax3.set_xticklabels(["2016","2017","2018","2019","2020","2021"])
for tick in [1.053,1.10,1.15,1.20,1.247]:
    ax3.axhline(y=tick,color="grey",linewidth=0.5)
ax3.axhline(mean_pre_corona,xmin=0.045,xmax=0.726,color="blue",linewidth=2,alpha=0.5)
ax3.axhline(mean_corona,xmin=0.726,xmax=0.955,color="red",linewidth=2,alpha=0.5)
ax3.text(16850,1.28,"Average rate of the pre-corona and corona time with means as horizontal lines",weight="bold",color="green")

### Adding a signature
ax3.text(16600,1,"©Bastian Hartmann"+" "*95+"Source: European Central Bank", color="#f0f0f0",backgroundcolor="#4d4d4d",size=14)

plt.show()
```
