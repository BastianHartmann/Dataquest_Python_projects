# Predicting Car Prices

The data set we will be working with contains information on various cars. For each car we have information about the technical aspects of the vehicle such as the motor's displacement, the weight of the car, the miles per gallon, how fast the car accelerates, and more. More information about the data set can be found <a href=https://archive.ics.uci.edu/ml/datasets/automobile>here</a> and it can be directly downloaded from <a href=https://archive.ics.uci.edu/ml/datasets/automobile>here.</a>

***
### Loading the Dataset

First, we load the data set into the dataframe `cars`.

```python
import pandas as pd

cars = pd.read_csv("additional_files/imports-85.data",names=["symboling","normalized_losses","make","fuel_type","aspiration","num_doors","body_style","drive_wheels","engine_location","wheel_base","length","width","height","curb_weight","engine_type","num_cylinders","engine_size","fuel_system","bore","stroke","compression_ratio","horsepower","peak_rpm","city_mpg","highway_mpg","price"])
```

```python
cars.info()
```

```python
cars.head(5)
```

***
### Data Cleaning

Based on the data set preview from the last section, we can tell that the `normalized-losses` column contains missing values represented using `"?"`. We replace these values and look for the presence of missing values in other numeric columns.

```python
import numpy as np

cars = cars.replace("?",np.nan)
cars.head(5)
```

```python
cars["price"].value_counts(dropna=False)
```

```python
cars["normalized_losses"] = cars["normalized_losses"].astype(float)
cars["bore"] = cars["bore"].astype(float)
cars["stroke"] = cars["stroke"].astype(float)
cars["horsepower"] = cars["horsepower"].astype(float)
cars["peak_rpm"] = cars["peak_rpm"].astype(float)
cars["price"] = cars["price"].astype(float)

cars.info()
```

After we converted the suitable columns to float numbers, we extract the columns containing only numerical values to the dataframe `numeric_cars`.

```python
continuous_values_cols = ['normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_size', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
numeric_cars = cars[continuous_values_cols]
numeric_cars.info()
```

Next, we analize how many null values remain in the columns of `numeric_cars`:

```python
numeric_cars.isnull().sum()
```

The above summary shows us that some values are missing in the columns `normalized_losses`,`bore`,`stroke`,`horsepower`,`peak_rpm` and `price`.</br>Because the `price` column is our prediction target, we drop all rows with missing values in this column.

```python
numeric_cars = numeric_cars.dropna(subset=["price"])
```

```python
numeric_cars.isnull().sum()
```

Because the remaining number of missing values is comparably low, we simply replace them with the mean value of the corresponding column.

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
```

```python
numeric_cars.isnull().sum()
```

```python
numeric_cars.head()
```

Finally, we normalize all feature columns so all values range from 0 to 1.

```python
norm_cars = (numeric_cars-numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
norm_cars["price"] = numeric_cars["price"]
```

```python
norm_cars.head()
```

***
### Univariate Model

Let's start with some univariate k-nearest neighbors models.<br>Starting with simple models before moving to more complex models helps us structure our code workflow and understand the features better.

Therefore, we first create a function, named `knn_train_test` that encapsulates the training and simple validation process. This function has 3 parameters -- training column name, target column name, and the dataframe object.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(feature_col,target_col,df):
    # Randomization
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    # Split df in train and test set
    split_num = int(len(rand_df)/2)
    train_df = rand_df.iloc[:split_num]
    test_df = rand_df.iloc[split_num:]
    
    # Fit KNN model with default values (k=5)
    knn = KNeighborsRegressor()
    knn.fit(train_df[[feature_col]],train_df[target_col])
    
    # Make predictions of test set
    predictions = knn.predict(test_df[[feature_col]])
    
    # Validate predictions
    rmse = mean_squared_error(test_df[target_col],predictions)**.5
    return rmse
```

Now we use this function to train and test univariate models using the different numeric columns in the data set.

```python
features = ['normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_size', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg']
target = "price"

rmses = {}
for feature in features:
    rmse = knn_train_test(feature,target,norm_cars)
    rmses[feature] = rmse
    
rmse_results = pd.Series(rmses)
rmse_results.sort_values()
```

The resulting *root mean squared error* (RMSE) values indicates that the column `engine_size` performed best for predicting the price for k=5.

To additionally analize the effect of different k values on the prediction performance, we modify the `knn_train_test` function to accept a parameter for the `k` value.<br>
This new function is called `knn_train_test_k_var`.

def knn_train_test_k_var(feature_col,target_col,df,k):
    # Randomization
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    # Split df in train and test set
    split_num = int(len(rand_df)/2)
    train_df = rand_df.iloc[:split_num]
    test_df = rand_df.iloc[split_num:]
    
    # Fit KNN model
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[[feature_col]],train_df[target_col])
    
    # Make predictions of test set
    predictions = knn.predict(test_df[[feature_col]])
    
    # Validate predictions
    rmse = mean_squared_error(test_df[target_col],predictions)**.5
    return rmse

For each feature we now vary `k` (`1`,`3`,`5`,`7` and `9`) and summarize the resulting prediction performance.

```python
k_vals = [1,3,5,7,9]

rmse_k_results = {}
for k in k_vals:
    rmses = {}
    for feature in features:
        rmse = knn_train_test_k_var(feature,target,norm_cars,k)
        rmses[feature] = rmse
    rmse_results = pd.Series(rmses)
    rmse_k_results[k] = rmse_results
    
rmse_results_df = pd.DataFrame(rmse_k_results)
```

```python
rmse_results_df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

rmse_results_df.T.plot(kind="line",figsize=(8,6))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel("k value")
plt.ylabel("RMSE")
plt.xticks(k_vals)
plt.show()
```

***
### Multivariate Model

We now further modify the `knn_train_test_k_var` function to accept a list of column names (instead of just a string) and rename it to `knn_train_test_k_var_multi`.

```python
def knn_train_test_k_var_multi(feature_cols,target_col,df,k):
    # Randomization
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    # Split df in train and test set
    split_num = int(len(rand_df)/2)
    train_df = rand_df.iloc[:split_num]
    test_df = rand_df.iloc[split_num:]
    
    # Fit KNN model
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[feature_cols],train_df[target_col])
    
    # Make predictions of test set
    predictions = knn.predict(test_df[feature_cols])
    
    # Validate predictions
    rmse = mean_squared_error(test_df[target_col],predictions)**.5
    return rmse
```

We know want to use the best features from the previous step. The best performing features for k=5 were:

```python
# best 5 features for k=5
k5_rmse = rmse_results_df[5]
k5_rmse.sort_values().iloc[0:5]
```

Next, we analize the prediction performance for the 2, 3, 4 and 5 best performing features using k=5.

```python
best2 = list(k5_rmse.sort_values().index[0:2])
best3 = list(k5_rmse.sort_values().index[0:3])
best4 = list(k5_rmse.sort_values().index[0:4])
best5 = list(k5_rmse.sort_values().index[0:5])

feature_lists = [best2,best3,best4,best5]

k5_rmse_results = {}
k = 5
for n_features in range(2,6):
    features = feature_lists[n_features-2]
    rmse = knn_train_test_k_var_multi(features,"price",norm_cars,k)
    k5_rmse_results["{}_best_features".format(n_features)] = {k:rmse}
```

```python
k5_rmse_df = pd.DataFrame(k5_rmse_results)
k5_rmse_df.index.name = "k-value"
k5_rmse_df
```

For k=5, the lowest RMSE was achieved using only the best 2 features.

***
### Hyperparameter Tuning

To additionally analyze the influence of `k` we now vary this hyperparameter value from `1` to `25` and plot the resulting RMSE values.

```python
k_vals = [x for x in range(1,26)]

k_rmse_results = {}
for n_features in range(2,6):
    k_dict = {}
    for k in k_vals:
        features = feature_lists[n_features-2]
        rmse = knn_train_test_k_var_multi(features,"price",norm_cars,k)
        k_dict[k] = rmse
    k_rmse_results["{} best features".format(n_features)] = k_dict
```

```python
k_rmse_df = pd.DataFrame(k_rmse_results)
k_rmse_df.index.name = "k value"
k_rmse_df
```

```python
k_rmse_df.plot(figsize=(8,6))
plt.ylabel("RMSE")
plt.show()
```

The plot reveals that the prediction performs best with eather the 2-best or 4-best features and a low `k` value of **2**. The increasing of the RMSE result with increasing features and/or higher `k` values indicates overfitting.
